## Choosing the Right Classification Model. Either malignant or Begin cancer

```python
import pandas as pd 
cancer=pd.read_csv("z.Data/Cancer_Data.csv")
cancer.head()
```

```python
x=cancer.drop(["Unnamed: 32", "diagnosis"], axis=1) 
y=cancer["diagnosis"] # We will predict this output
```

```python
## Training Phase
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

### Making the Varience fixed  
- In the Features there may be difference varience leading 
different mean so training data may results error output

```python
from sklearn.preprocessing import StandardScaler
# Scale the features
scaler = StandardScaler()
x_train_scl = scaler.fit_transform(x_train)
x_test_scl = scaler.transform(x_test)
```

### Using Different Model to see the classification's score result.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

models = {
    "Logistic Reg": LogisticRegression(max_iter=10000, random_state=42),
    "Linear SVC": LinearSVC(random_state=42),
    "Rand Forest": RandomForestClassifier(random_state=42)
}

# Evaluating each model
results = {}
for name, model in models.items():
    model.fit(x_train_scl, y_train)
    
    # Predict on test data
    y_pred = model.predict(x_test_scl)
    
    # Calculate metrics on TEST data
    train_score = model.score(x_train_scl, y_train)  # Training accuracy
    test_score = model.score(x_test_scl, y_test)     # Test accuracy
    
    accuracy = accuracy_score(y_test, y_pred) 
    precision = precision_score(y_test, y_pred, pos_label='M') 
    recall = recall_score(y_test, y_pred, pos_label='M')
    f1 = f1_score(y_test, y_pred, pos_label='M') # M as positive 
    
    results[name] = {
        "Train_Accuracy": train_score,    # Useful to detect overfitting
        "Test_Accuracy": test_score,      # Same as accuracy below
        "Accuracy": accuracy,
        "Precision": precision,
        "Recall": recall,
        "F1-Score": f1
    }

# Displaying results as whole
import pandas as pd
results_df = pd.DataFrame(results).T
print(results_df)
```
