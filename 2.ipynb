## Training Model when Dataset contains missing values

```python
import pandas as pd
import numpy as np 
covid=pd.read_csv("z.Data/covid_data.csv")
covid.loc[2:4, "New_Cases"]=np.nan
covid.loc[[0,3], "Recovered"]=np.nan
covid.head() 
```

### let check how much missing data,

```python
covid.isna().sum()
```

### Now let put value on those column 

```python
covid["New_Cases"].fillna(covid["New_Cases"].mean(), inplace=True)
covid["Recovereds"].fillna(covid["Recovered"].mean(), inplace=True)
```

```python
x= covid.drop("New_Deaths", axis=1) 
y= covid["New_Deaths"]
```

```python
from sklearn.model_selection import train_test_split
x=pd.get_dummies(x,drop_first=True)  
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import RandomForestRegressor 
clf= RandomForestRegressor() 
clf.fit(x_train, y_train)  
```
