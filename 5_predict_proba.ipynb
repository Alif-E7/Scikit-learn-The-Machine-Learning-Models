```python
import pandas as pd 
cancer=pd.read_csv("z.Data/Cancer_Data.csv")
cancer.head(10) 
```

```python

from sklearn.model_selection import train_test_split
x=cancer.drop(["Unnamed: 32", "diagnosis"], axis=1) 
y= cancer["diagnosis"]
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.2,random_state=17)
```

- Oh i forgot before pred i need to scaler 

```python
from sklearn.preprocessing import StandardScaler
# Scale the features
scaler = StandardScaler()
x_train_scl = scaler.fit_transform(x_train)
x_test_scl = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression 
model= LogisticRegression(max_iter=500, random_state=17) 
model.fit(x_train_scl, y_train) 
```

#### Now predict... 

```python
y_test.head(10)
```

```python
model.predict(x_test_scl[:20])
```

## Let calculate another radius_mean

```python

from sklearn.model_selection import train_test_split
cancer=pd.get_dummies(cancer, drop_first=True)
x=cancer.drop("Unnamed: 32", axis=1) 
y= cancer["radius_mean"]
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.25,random_state=17) 
```

```python
from sklearn.ensemble import RandomForestRegressor
mdl=RandomForestRegressor(random_state=17) 
mdl.fit(x_train, y_train)
```

```python
ypredR= mdl.predict(x_test)
ypredR
```

```python
y_test
```

```python
from sklearn.metrics import mean_absolute_error 
mean_absolute_error(y_test, ypredR) # On average error 
```
