## 1. Separate Numaric and Text 

```python
categorical_features = ['Brand',	'Model', 'Color','Fuel_Type' ,'Transmission'] 
numerical_features = ['Year', 'Price', 'Mileage', 'Car_ID', 'Owner_Type_Third']
```

## 2. Now  convert all together 

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler

prework = ColumnTransformer(transformers=[
    ('num', Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', StandardScaler())
    ]), numerical_features),

    ('cate', Pipeline(steps=[
        ('fillna', SimpleImputer(strategy='most_frequent')),
        ('tonum', OneHotEncoder(handle_unknown='ignore'))
    ]), categorical_features)
])
```
