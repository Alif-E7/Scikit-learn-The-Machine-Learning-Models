## We want to train a model clf then using clf try to predict unkmown data

```python
import pandas as pd 
covid=pd.read_csv("z.Data/covid_data.csv")
covid.sample(frac=0.07, random_state=42)
```

### Let we want to predicts New_Deaths after trainging model with other data

```python
# Keep eveything without New_deaths bcz we want to predict it 
x= covid.drop("New_Deaths", axis=1) # axis=1 -> 
x.head()
```

```python
y=covid["New_Deaths"]
y.head()
```

```python
from sklearn.model_selection import train_test_split 
x = pd.get_dummies(x, drop_first=True)  # Object to Numaric
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=42)
x.head()
```

```python

from sklearn.ensemble import RandomForestRegressor 
model= RandomForestRegressor() 
model.fit(x_train, y_train) # For this X == this Y 
model.score(x_train, y_train)
```
