```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

# 1. Load Data
covid = pd.read_csv("z.Data/covid_data.csv")

# 2. Define Features (X) and Target (y)
# Drop ONLY the target column. Keep Date and Country in X for now.
x = covid.drop("New_Deaths", axis=1) 
y = covid["New_Deaths"]

# 3. Convert ALL Object Columns to Numeric at once
# This keeps numerical columns (Total_Cases) AND converts text (Country, Date)
x = pd.get_dummies(x, drop_first=True) 

# 4. Split into Train and Test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=42)

# 5. Verify the result
print(f"Columns in X: {x.columns.tolist()}")
print(f"Shape: {x.shape}")

# 6. Train Model
model = RandomForestRegressor()
model.fit(x_train, y_train)
print(f"Score: {model.score(x_train, y_train)}")
```
