## Choosing the Right Classification Model. Either malignant or Begin cancer

```python
import pandas as pd 
cancer=pd.read_csv("z.Data/Cancer_Data.csv")
cancer
```

### -Same as before filling missing data

```python
x=cancer.drop(["Unnamed: 32", "diagnosis"], axis=1) 
y=cancer["diagnosis"] # We will predict this output
```

```python
## Training Phase
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score


# Scale the features
scaler = StandardScaler()
x_train_scl = scaler.fit_transform(x_train)
x_test_scl = scaler.transform(x_test)

# Re-initialize models
models = {
    "Logistic Regression": LogisticRegression(max_iter=10000, random_state=42),
    "Linear SVC": LinearSVC(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42)
}

# Evaluate each model
results = {}
for name, model in models.items():
    model.fit(x_train_scl, y_train)
    y_pred = model.predict(x_test_scl)

    # Calculate metrics
    score= model.score(x_train_scl, y_train)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, pos_label='M')
    recall = recall_score(y_test, y_pred, pos_label='M')
    f1 = f1_score(y_test, y_pred, pos_label='M')

    results[name] = {
        "Accuracy": accuracy,
        "Precision": precision,
        "Recall": recall,
        "F1-Score": f1,
        name:score 
    }

# Display results
results
```
