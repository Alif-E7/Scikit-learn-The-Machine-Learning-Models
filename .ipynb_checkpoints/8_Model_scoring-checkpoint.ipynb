```python
import pandas as pd
import numpy as np 
covid=pd.read_csv("z.Data/covid_data.csv") 
covid.head()
```

```python
from sklearn.model_selection import train_test_split
x=covid.drop(columns=["Date", "Country", "New_Deaths"] )
y=covid["New_Deaths"] 

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
num_cols = [
    'Total_Cases',
    'New_Cases',    
    'Total_Deaths',
    'Recovered',
    'Active_Cases',
    'Serious_Critical',
    'Tests_Per_Million',
    'Population',
] 
cat=['Alert']

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score

preprocessor = ColumnTransformer(transformers=[ 
    ('num', Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='median')), 
        ('scaler', StandardScaler())                   
    ]), num_cols),
    
    ('cat', Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='most_frequent')), 
        ('onehot', OneHotEncoder(handle_unknown='ignore'))
         #- Converts categorical variables into binary (0/1) columns.
    ]), cat)
])
```

```python
from sklearn.ensemble import RandomForestRegressor
preprocessor= RandomForestRegressor(n_estimators=100 ,random_state=42)
                    
preprocessor.fit(X_train, y_train) 

preprocessor.score(X_test, y_test)
```
